```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salary_and_number_of_childeren_data.csv')
num = df.to_numpy()
x, y = num[:, 3], num[:,2]

#df.plot.scatter(x='Salery', y='Number of Childrn')
```

```python
x
```

```python
plt.scatter(x, y)
plt.show()
```

```python
scaled_x = []
scaled_y = []

for i in x:
    scaled_x.append(i/5)

for i in y:
    scaled_y.append(i/200000)

scaled_x = np.array(scaled_x)
scaled_y = np.array(scaled_y)

scaled_x
```

```python
b = 0.2
w = 1

plt.scatter(scaled_x, scaled_y)

axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + w * x_vals
plt.plot(x_vals, y_vals, '--', color='red')

plt.show()
```

```python
x_vals
```

```python
def compute_cost(x, y, w, b): 

    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
compute_cost(scaled_x , scaled_y, w, b)
```

```python
w = np.linspace(0, 10, 100)
```

```python
j = []

for i in w:
    j.append(compute_cost(scaled_x , scaled_y, i, b))
             
j = np.array(j)

j
```

```python
plt.plot(w, j)
plt.show()
```
